```python
from transformers import pipeline
import pandas as pd

# Load dataset
df = pd.read_csv('/content/training.1600000.processed.noemoticon.csv.zip', encoding='ISO-8859-1', header=None)
df.columns = ['target', 'ids', 'date', 'flag', 'user', 'text']

# Sample a few tweets
sample_df = df[['text']].sample(5, random_state=42).reset_index(drop=True)

# Load a sentiment-analysis pipeline (offline, DuckDuckGo-like style)
classifier = pipeline("sentiment-analysis")

# Run sentiment predictions
sample_df['predicted_sentiment'] = sample_df['text'].apply(lambda x: classifier(x)[0]['label'].lower())

# Optional: convert to binary label
sample_df['predicted_label'] = sample_df['predicted_sentiment'].map({'negative': 0, 'positive': 1})

print(sample_df)
```
